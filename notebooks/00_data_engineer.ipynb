```python
#--------------------------------------------------
# Step 1: Import required libraries
#--------------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#--------------------------------------------------
# Step 2: Load the dataset
#--------------------------------------------------

print("Loading dataset...")
df = pd.read_csv('/workspaces/End_to_end_insurance_price_predection/data/raw/insurance.csv')
print("Dataset loaded successfully.")
```

```python
# -------------------------------------------------------------
# Step 3: Understand structure ‚Äî schema, datatypes, nulls
# -------------------------------------------------------------
print("\nüìä Data types and null value check:")
print(df.info())

print("\n‚ùì Any missing values?")
print(df.isnull().sum())
```

```python
#--------------------------------------------------
# Explore statistical summary of numerical columns
#--------------------------------------------------
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("\nüîç Statistical summary of numerical columns:")
print(df[numerical_cols].describe())
```

```python
#--------------------------------------------------
# Step 3:Standardize column naemes
#--------------------------------------------------
print("\nüî§ Standardizing column names...")
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
print("Column names standardized.")
print(df.columns.tolist())
```

```python
#--------------------------------------------------
# Check for duplicate rows
#--------------------------------------------------
duplicates = df.duplicated().sum()
print(f"\nüîç Number of duplicate rows: {duplicates}")
if duplicates > 0:
    print("Removing duplicate rows...")
    df = df.drop_duplicates()
    print("Duplicate rows removed.")
```

```python
#--------------------------------------------------
# Save the cleaned version
#--------------------------------------------------
print("\nüíæ Saving cleaned dataset...")
df.to_csv('/workspaces/End_to_end_insurance_price_predection/data/processed/insurance_cleaned.csv', index=False)
print("Cleaned dataset saved successfully.")
```
